```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
prosaico = direcao[direcao["ssid"]=="Prosaico01"]
fig = sns.boxplot(x="DIRECAO",y="rssi",data=prosaico)
fig.set_xlabel("Direções")
fig.set_ylabel("RSSI (dbm)")
fig.set_title("Rede Prosaico01")
fig = fig.get_figure()
fig.savefig("prosaico01_direcoes.png")
```

```python
testerssi = direcao[direcao["ssid"]=="testerssi"]
fig = sns.boxplot(x="DIRECAO",y="rssi",data=testerssi)
fig.set_xlabel("Direções")
fig.set_ylabel("RSSI (dbm)")
fig.set_title("Rede testerssi - Gerado pelo ESP32")
fig = fig.get_figure()
fig.savefig("testerssi_direcoes.png")
```

```python
devices = pd.read_excel("RSSI_DIREÇÕES.xlsx", sheet_name="devices")
```

```python
# sns.figure(figsize=(20,10))
plt.figure(figsize=(15,5))
fig = sns.stripplot(x="x",y="FRENTE",data=devices)
fig.set_xlabel("Amostras")
fig.set_ylabel("RSSI (dbm)")
fig.set_title("Rede Prosaico01 - 1 ESP32 coletando em 4 direções")
fig.annotate("FRENTE", 
            xy=(30, -50),
            xycoords='data',
            xytext=(10, -48),
            textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
fig.annotate("ESQ", 
            xy=(150, -53),
            xycoords='data',
            xytext=(150, -57),
            textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
fig.annotate("TRÁS", 
            xy=(250, -53),
            xycoords='data',
            xytext=(250, -57),
            textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
fig.annotate("DIR", 
            xy=(350, -54),
            xycoords='data',
            xytext=(350, -51),
            textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
fig = fig.get_figure()
fig.savefig("prosaico01_onedevice.png")
```
