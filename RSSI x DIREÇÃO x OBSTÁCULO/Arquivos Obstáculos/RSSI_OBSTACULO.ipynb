```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
obstaculo = pd.read_excel("RSSI_OBSTACULO.xlsx", sheet_name="TODOS_FILTRADOS")
```

```python
obstaculo.head()
```

```python
testerssi = obstaculo[obstaculo["ssid"]=="testerssi"]
fig = sns.boxplot(x="OBSTACULO",y="rssi",data=testerssi, hue="DIRECAO")
fig.set_xlabel("OBSTACULO")
fig.set_ylabel("RSSI (dbm)")
fig.set_title("Rede testerssi - Gerado pelo ESP32")
# To relocate the legend
fig.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
fig = fig.get_figure()
fig.savefig("testerssi_obstaculos.png")
```
