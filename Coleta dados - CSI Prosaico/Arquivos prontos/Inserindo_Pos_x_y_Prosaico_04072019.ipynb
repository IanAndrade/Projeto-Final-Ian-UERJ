```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
csidf = pd.read_excel("subcarriers_amp_pos_pot.xlsx", sheet_name="subcarriers_amp_pos_pot")
csidf.drop("Unnamed: 0", axis=1, inplace=True)
csidf.head()
```

```python
for x in csidf.index:
    if csidf["local"].loc[x]=="A1":
        csidf['x'][x] = 1
        csidf['y'][x] = 6.05
    if csidf["local"].loc[x]=="A2":
        csidf['x'][x] = 0.7
        csidf['y'][x] = 4.12
    if csidf["local"].loc[x]=="A3":
        csidf['x'][x] = 0.7
        csidf['y'][x] = 1.2
    if csidf["local"].loc[x]=="B1":
        csidf['x'][x] = 2.15
        csidf['y'][x] = 6.05
    if csidf["local"].loc[x]=="B2":
        csidf['x'][x] = 2.55
        csidf['y'][x] = 4.12
    if csidf["local"].loc[x]=="B3":
        csidf['x'][x] = 2.55
        csidf['y'][x] = 1.2
    if csidf["local"].loc[x]=="D1":
        csidf['x'][x] = 4.85
        csidf['y'][x] = 5.8
    if csidf["local"].loc[x]=="D2":
        csidf['x'][x] = 4.85
        csidf['y'][x] = 4
    if csidf["local"].loc[x]=="D3":
        csidf['x'][x] = 4.85
        csidf['y'][x] = 1.2
    if csidf["local"].loc[x]=="E1":
        csidf['x'][x] = 7.35
        csidf['y'][x] = 6.55
    if csidf["local"].loc[x]=="E2":
        csidf['x'][x] = 7.35
        csidf['y'][x] = 4.12
    if csidf["local"].loc[x]=="E3":
        csidf['x'][x] = 7.35
        csidf['y'][x] = 1.5
```
