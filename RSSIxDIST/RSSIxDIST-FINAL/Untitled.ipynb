```python
g = uerj[uerj["device"]=="ESQ"]["roda_distance"].unique()

for x in g:
    print("####")
    m = uerj[(uerj["device"]=="ESQ") & (uerj["roda_distance"]==x)]["rssi"].mean()
    d = uerj[(uerj["device"]=="ESQ") & (uerj["roda_distance"]==x)]["rssi"].std()
    maximo = uerj[(uerj["device"]=="ESQ") & (uerj["roda_distance"]==x)]["rssi"].max()
    minimo = uerj[(uerj["device"]=="ESQ") & (uerj["roda_distance"]==x)]["rssi"].min()
    moda = uerj[(uerj["device"]=="ESQ") & (uerj["roda_distance"]==x)]["rssi"].mode()
    print("Media RSSI distancia =" + str(x) + ":"+ str(m))
    print("Desvio padrão distancia = " + str(x) + ":"+ str(d))
    print("Máximo RSSI distancia = " + str(x) + ":"+ str(maximo))
    print("Min RSSI distancia = " + str(x) + ":"+ str(minimo))
    print("Moda RSSI distancia =  " + str(x) + ":"+ str(moda))
```
